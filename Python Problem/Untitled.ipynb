```python
def get_tweets_Json(username): 
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token_key, access_token_secret)
    api = tweepy.API(auth)
    tweets = api.user_timeline(screen_name=username,tweet_mode="extended", count=100)
    json_tweets = [tweet._json for tweet in tweets]
    with open('Json_Tweet_File.txt', 'wt') as outfile:
       json.dump(json_tweets, outfile)        
    return 'Json_Tweet_File.txt'
    
    
```
